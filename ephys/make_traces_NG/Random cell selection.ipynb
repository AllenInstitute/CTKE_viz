```python
import pandas as pd
import numpy as np
import os
```

```python
metadata_df = pd.read_csv("../m1_patchseq_meta_data.csv", sep="\t", index_col=0)
metadata_df
```

```python
metadata_df["RNA type"].value_counts()
```

```python
metadata_df.groupby("RNA type").sample(n=3, replace=False)
```

```python
chosen_list = []
for ttype in metadata_df["RNA type"].dropna().unique():
    ttype_df = metadata_df.loc[metadata_df["RNA type"] == ttype, :]
    if ttype_df.shape[0] <= 3:
        chosen_list.append(ttype_df)
    else:
        chosen_list.append(ttype_df.sample(n=3, replace=False, random_state=615))
```

```python
sampled_df = pd.concat(chosen_list)
```

```python
sampled_df
```

```python
filepaths = []
for i, r in sampled_df.iterrows():
    cell_path = os.path.join("/allen/programs/celltypes/workgroups/ivscc/nathang/cell_cards_ephys",
                             "nwb",
                             "000008",
                             f'sub-{r.Mouse.replace("_", "-")}',
                             f'sub-{r.Mouse.replace("_", "-")}_ses-{r.Cell.replace("_", "-")}_slice-{r.Slice.replace("_", "-")}_cell-{r.Cell.replace("_", "-")}_icephys.nwb'
                            ) 
    if os.path.exists(cell_path):
        filepaths.append(cell_path)
    else:
        print("File not found")
        print(i, r["RNA type"], cell_path)
        filepaths.append("")
```

```python
sampled_df["nwb_filepath"] = filepaths
sampled_df = sampled_df.loc[sampled_df["nwb_filepath"] != "", :]
sampled_df
```

```python
sampled_df.to_csv("../randomly_selected_cells_for_ttypes.csv")
```
