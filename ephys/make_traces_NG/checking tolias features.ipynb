```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
%matplotlib inline
```

```python
features_df = pd.read_csv("../m1_patchseq_ephys_features.csv", index_col=0)
```

```python
meta_df = pd.read_csv("../m1_patchseq_meta_data.csv", sep="\t", index_col=0)
```

```python
feat_type_df = features_df.join(meta_df.loc[:, ["Cell", "RNA type"]].set_index("Cell"))
```

```python
feat_type_df
```

```python
for c in features_df.columns:
    print(c)
```

```python
plt.figure(figsize=(5, 5))
sns.violinplot(data=feat_type_df.dropna().loc[feat_type_df.dropna()["RNA type"].str.startswith("L5 ET"), :],
              x="RNA type",
              y="Upstroke-to-downstroke ratio",
              cut=0,
              scale="width",
              inner=None,
              order=["L5 ET_1", "L5 ET_2", "L5 ET_3", "L5 ET_4"],
             )


sns.swarmplot(data=feat_type_df.dropna().loc[feat_type_df.dropna()["RNA type"].str.startswith("L5 ET"), :],
              x="RNA type",
              y="Upstroke-to-downstroke ratio",
              color="black",
              order=["L5 ET_1", "L5 ET_2", "L5 ET_3", "L5 ET_4"],              
             )
```
